### Exercício 02 – Manipulando dados com Pandas

a) Crie um programa em pandas para ler o arquivo em Excel imobiliária.xlsx

e) Escreva um programa para renomear o cabeçalho da tabela com os seguintes nomes:

```python
df = df.rename(columns={
    'Coluna 1':'Filial',
    'Coluna 2':'Vendedor',
    'Coluna 3':'Data',
    'Coluna 4':'Tipo',
    'Coluna 5':'Área(m²)',
    'Coluna 6':'Dorms.',
    'Coluna 7':'Garag.',
    'Coluna 8':'Piscina',
    'Coluna 8.1':'Preço',
    'Coluna 10':'Comissão'
})

df.head()
```

g) Quais são os tipos de dados que temos no dataframe?

h) Há valores faltantes ou nulos?

```python
df.info()
```

i) Calcule a soma, média, máximo, mínimo e contagem da coluna Preço.

```python
Soma = df['Preço'].sum()
print(f'A soma dos preços é:',Soma)

Média = df['Preço'].median()
print(f'A média dos preços é:',Média)

Máximo = df['Preço'].max()
print(f'O maior preço é:',Máximo)

Mínimo = df['Preço'].min()
print(f'O menor preço é:',Mínimo)

Contagem = df['Preço'].count()
print(f'A quantidade de preços informados é:',Contagem)
```
